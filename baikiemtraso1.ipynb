```python
import pandas as pd
######c1
names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivan", "Jack", "Kelly", "Liam", "Mona", "Nina", "Oscar"]
ages = [25, 30, 35, 28, 22, 45, 34, 31, 27, 29, 33, 40, 26, 32, 36]
salary = [50000, 60000, 70000, 55000, 52000, 80000, 72000, 68000, 61000, 59000, 63000, 77000, 53000, 66000, 75000]

data = {
    'Name': names,
    'Age': ages,
    'Salary': salary
}

df = pd.DataFrame(data)
#######cau 2 
print(df)
```

```python
#3
filtered_df = df[df['Age'] > 28]
print(filtered_df)
```

```python
#4
df2 = df['Salary'].mean()
print("Giá trị trung bình của cột Salary là:", df2)
```

```python
#5
group_df = df.groupby('Age')['Salary'].sum().reset_index()
print(group_df)
```

```python
#6
df3_df = df.sort_values(by='Salary', ascending=False)
print(df3_df)
```

```python
####7
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Age'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of Individuals')
plt.xticks(rotation=45) 
plt.show()
```

```python
#8:vẽ biểu đồ dường cột cho salary 
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Salary'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary of Individuals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#9:
plt.figure(figsize=(8, 8))
plt.pie(df['Age'], labels=df['Name'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Age Distribution Among Individuals')
plt.axis('equal')  # Đảm bảo biểu đồ đường tròn được vẽ tròn
plt.show()
```

```python
#10
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='b', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot of Age vs Salary')
##Thêm nhãn cho từng điểm dữ liệu
for i in range(len(df)):
    plt.text(df['Age'][i], df['Salary'][i], df['Name'][i], fontsize=9, ha='right')

plt.grid(True)
plt.show()
```

```python
#c11 
has_nan = df.isna().any().any()
print("DataFrame có giá trị NaN:", has_nan)
```

```python
#C12 
#tính giá trị trung bình của cột Age 
mean_age = df['Age'].mean()
#Thay thế các giá trị của cột 'Age' lớn hơn 30 bằng giá trị trung bình
df.loc[df['Age'] > 30, 'Age'] = mean_age
print(df)
```

```python
#c13
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
print(df)
```

```python
#c14    
bins = [0, 30, 45, 100]
labels = ['Young', 'Middle-aged', 'Old']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

print(df)
```

```python
#c15:tính toán tỷ lệ phần trăm 
df['Salary_change_pct'] = df['Salary'].pct_change() * 100
print(df.to_string(index=False))
```
